```python
payload= {
    'action' : 'login',
    'username': os.environ.get("KAGGLE_USERNAME"),
    'password': os.environ.get("KAGGLE_PASSWORD")
}

def extract_data(url, file_path):
    with session() as c:
        c.post('https://www.kaggle.com/account/login', data=payload)
        print('file_path in def_extract:',file_path)
        with open(file_path, 'w') as handle:
            response = c.get(url, stream=True)
            for block in response.iter_content(1024):
                handle.write(block)
```

```python
train_url = 'https://www.kaggle.com/c/titanic/download/train.csv'
test_url = 'https://www.kaggle.com/c/titanic/download/test.csv'

print(os.path)

raw_data_path = os.path.join(os.path.pardir, 'data', 'raw')
print(raw_data_path)
train_data_path = os.path.join(raw_data_path, 'train.csv')
print(train_data_path)
test_data_path = os.path.join(raw_data_path, 'test.csv')
print(test_data_path)


#Extract data
extract_data(train_url, train_data_path)
extract_data(test_url, test_data_path)
```

```python
# A Reusable script to extact data

get_raw_data_script_file = os.path.join(os.path.pardir, 'src', 'data', 'get_raw_data.py')
print(get_raw_data_script_file)

import os
import logging
from dotenv import find_dotenv, load_dotenv
from requests import session


#String to raw
def str_to_raw(s):
    raw_map = {8:r'\b', 7:r'\a', 12:r'\f', 10:r'\n', 13:r'\r', 9:r'\t', 11:r'\v'}
    return r''.join(i if ord(i) > 32 else raw_map.get(ord(i), i) for i in s)


#payload
payload ={
    'action': 'login',
    'username': os.environ.get("KAGGLE_USERNAME"),
    'password': os.environ.get("KAGGLE_PASSWORD")
}


def extract_data1(url, file_path):
    with session() as c:
        c.post('https://www.kaggle.com/account/login', data=payload)
        file_path_abs = os.path.abspath(file_path)
        with open(file_path, 'w') as handle:
            response = c.get(url, stream=True)
            for block in response.iter_content(1024):
                handle.write(block)
```

```python
from pathlib import Path
def main(project_dir):
    #Logger
    logger = logging.getLogger(__name__)
    logger.info('getting raw data')
    
    #url info
    train_url = 'https://www.kaggle.com/c/titanic/download/train.csv'
    test_url = 'https://www.kaggle.com/c/titanic/download/test.csv'
    
    #file paths
    #raw_data_path = os.path.join(project_dir, 'data', 'raw')
    raw_data_path = Path(project_dir)
    #train_data_path = os.path.join(raw_data_path, 'train.csv')
    #test_data_path = os.path.join(raw_data_path, 'test.csv')
    
    train_data_path = raw_data_path / "train.csv"
    test_data_path = raw_data_path / "test.csv"
    
    
    #Data extraction
    extract_data1(train_url, train_data_path)
    extract_data1(test_url, test_data_path)
    

if __name__ == '__main__':
    project_dir = os.path.join(os.path.dirname("__file__"), os.pardir, os.pardir)
    
    dotenv_path = find_dotenv()
    
    load_dotenv(dotenv_path)
    
    main(project_dir)
```
